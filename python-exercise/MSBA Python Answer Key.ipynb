Question 11: Which release year saw the lowest average user price?

```python
# Grouping Release Year and taking average of user prices 
# then arrangign values in ascending order and displaying the 
# lowest average price
data['LowUserPrice'].groupby(data.ReleaseYear).mean().sort_values().head(1)
```

Question 12: Reshape/ Pivot the data to create a dataframe that has SKUs in one column, all continuous numeric variables (StrengthFactor, PriceReg, ItemCount, LowUserPrice, LowNetPrice) as values in one column and all their respective numerical values in a 3rd column (Hint: this dataframe should have only 3 columns in total- SKU, Variable, Value) 

```python
# Goal is to have 1 column with all SKUs, one column containing all numerical variables and another with their respective values
# This is called a pivot in excel, melt/cast in R
# in Python as well, we can use the melt method of pandas
pd.melt(data, id_vars='SKU_number', value_vars=['StrengthFactor', 'PriceReg', 'ItemCount','LowUserPrice','LowNetPrice'])
```

Question 13: Create a chart/plot to show average lownetprice trend by release year for last 10 years.

```python
# Importing and setting matplotlib
%matplotlib inline
import matplotlib.pyplot as plt

# Plotting the chart
x = data.groupby('ReleaseYear')['LowNetPrice'].mean().tail(10)
x.plot(kind='bar')
```

Question 14: List the top 10 and bottom 10 SKUs by number of orders for both active and historical sales.

```python
# Active Sales Top 10 SKUs
active['ItemCount'].groupby(active.SKU_number).sum().sort_values(ascending=False).head(10)
```

```python
# Active Sales Bottom 10 SKUs
active['ItemCount'].groupby(active.SKU_number).sum().sort_values().head(10)
```

```python
# Historical Sales Top 10 SKUs
historical['ItemCount'].groupby(historical.SKU_number).sum().sort_values(ascending=False).head(10)
```

```python
# Historical Sales Bottom 10 SKUs
historical['ItemCount'].groupby(historical.SKU_number).sum().sort_values().head(10)
```

Question 15: Analyze the relationship between StrengthFactor and PriceReg (Hint: a chart/plot might help). Create 3 charts in total – one for the entire data, and one each for historical and active sales. 

```python
# Entire dataset
plt.style.use('seaborn-whitegrid')
plt.plot(data.PriceReg, data.StrengthFactor, 'o', color='black')
plt.xlabel('PriceReg')
plt.ylabel('StrengthFactor')
```

```python
# Active Sales 
plt.plot(active.PriceReg, active.StrengthFactor, 'o', color='black')
plt.xlabel('PriceReg')
plt.ylabel('StrengthFactor')
```
