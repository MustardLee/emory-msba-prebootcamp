```python
#12.Reshape/ Pivot the data to create a dataframe that has SKUs in one column, all continuous numeric variables (StrengthFactor, PriceReg,ItemCount, LowUserPrice, LowNetPrice) as values in one column and all their respective numerical values in a 3rd column (Hint: this dataframe should have only 3 columns in total- SKU, Variable, Value)
file_panel = file.melt(id_vars=['SKU_number'],value_vars=['StrengthFactor','PriceReg','ItemCount','LowUserPrice','LowNetPrice'],var_name=['Variable'],value_name='Value')
file_panel.rename(columns={'SKU_number':'SKU'})
```

```python
#13.Create a chart/plot to show average lownetprice trend by release year for last 10 years
LowUserPrice_for_10_years=pd.DataFrame(file.groupby("ReleaseYear")['LowNetPrice'].mean()).sort_values(by="ReleaseYear",ascending=False).head(10)
import seaborn as sns
import matplotlib.pyplot as plt

sns.relplot(x = 'ReleaseYear', y = 'LowNetPrice',data=LowUserPrice_for_10_years,kind='line')
plt.show()
```
